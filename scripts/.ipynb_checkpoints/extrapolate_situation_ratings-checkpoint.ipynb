```python
# data processing
import pandas as pd
import numpy as np
from pathlib import Path
from numpy import nanmean

# formatting
import re
def add_comma(match):
    return match.group(0) + ','

# statistical measures
from scipy.stats import pearsonr, mode 
from scipy import stats
from scipy.stats import zscore

# plots
import seaborn as sns
from matplotlib import pyplot as plt
from matplotlib.patches import Patch

import pandas as pd
import numpy as np
from pathlib import Path

from scipy.stats import pearsonr, mode 
from scipy import stats
from numpy import nanmean

from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import KFold

#Models
from sklearn.svm import SVR
from sklearn.linear_model import RidgeClassifier, Ridge
from sklearn.neighbors import KNeighborsClassifier, KNeighborsRegressor
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor
from sklearn.neural_network import MLPRegressor

# random state
randState = 0

# warning messages
import warnings
warnings.filterwarnings("ignore", message="An input array is constant; the correlation coefficient is not defined.")
```

```python
#random state
randState = 0

#functions to chose data set and embeddings type:
# Choose data set
def loadData(data = "RSQ", folder = "final", emb = "SENTENCEBERT"):
    # potentially add item embeddings to the situation embeddings
    embeddings = "_situations_embeddings_" + emb.upper() +  ".csv"
    save = "_" + emb.upper() +  ".csv"
    res = data.lower() + "_situations_text.csv"
    responses = pd.read_csv("../embeddings/" + folder + "/Situations/"+res, index_col=0).drop(["situation"], axis=1) #reversed
    embeddings_df = pd.read_csv("../embeddings/" + folder + "/Situations/"+data.lower() + embeddings, index_col=0)
    return responses, embeddings_df

#choose prediction model
def predModel(nr=1,par=1, randState=0):
    if nr == 0:
        model = Ridge(alpha=par, random_state=randState)
        modName = "Ridge"
    # elif nr == 1:
    #     model = SVR(C=par)
    #     modName = "SVR"
    elif nr == 1:  
        model = KNeighborsRegressor(n_neighbors=par)
        modName = "KnnReg"
    elif nr == 2:
        model = RandomForestRegressor(n_estimators=int(par), random_state=randState)
        modName = "RandomForest"
    elif nr == 3:
        model = GradientBoostingRegressor(n_estimators=int(par), random_state=randState)
        modName = "GradientBoosting"
    elif nr == 4:
        # Simple MLP with a fixed architecture, only varying the regularization parameter
        model = MLPRegressor(hidden_layer_sizes=(100,), alpha=par, random_state=randState)
        modName = "MLPRegressor"
    else:
        raise ValueError("Model number not recognized")
    
    return model, modName

def corrQuestionBased(x,y):
  #calculate correlations and p value for every person, then average:
  results = pd.DataFrame(np.full(shape=(x.shape[1],3),fill_value=0), index = x.columns, columns=["Correlation","p-value","L1 Loss"])
  
  #new function, much faster, #NOT FOR IPIP, #adjust for p value of averages
  corrVals = x.corrwith(y, 0, method=lambda x, y: np.round(pearsonr(x, y), 6)).fillna(0)
  results.iloc[:,:2] = corrVals.values.tolist()
  results.iloc[:,2] = np.absolute(x - y).sum(1)/x.shape[1] 
  results = results.fillna(0)
  t, pval_total = stats.ttest_1samp(results.Correlation, popmean=0) #get p-value of average correlation tested against 0, #change to make n decimals!
  mean, ci_lower, ci_upper = mean_confidence_interval(results.Correlation)
  return results, [results.Correlation.mean(), pval_total, ci_lower, ci_upper, t, results.shape[0]-2]

def getParams(m):
  #choose parameter range
    if m == 1:  # KNN and KNN Regression
        list_par = [1, 5, 10, 15, 50, 100]  # parameter search
    elif m in [2,3]:  # Random Forest and Gradient Boosting
        list_par = [10, 50, 100]  # number of trees in the forest/number of boosting stages
    elif m == 4:  # MLPRegressor
        list_par = [0.001, 0.01, 0.1, 1]  # L2 penalty (regularization term)
    else:  # Ridge, SVR
        list_par = [10**x for x in range(-3, 3)]  # parameter search, including lower values for Ridge and SVR
    
    return list_par

def compareModels(verbose = 0, folder = "final", ):
    for l in ['sentencebert']: #go through the three embeddings
      for k in range(2,5):                       # calculate for each prediction model: Ridge, 
          #model:
          m = k            #0: Ridge, #1: RidgeClass, #2:KNN, #3: Kernel SVM (RBF), #4: KNN regression
          
          #data:
          d = "rsq"       #data sets:   # BIG5, IPIP (all items), RIASEC, HSQ, 16PF
          e = l
          
          ############################ choose output here ##############################
          Path("../results/" + folder + "/Situations/" + d.upper()).mkdir(parents=True, exist_ok = True)
          #####################################################################################

          #choose data set, embeddings, encoding
          responses, data_s = loadData(data = d, folder = folder, emb = e)
        
          # drop all rows of responses that have is not interpretable as a number and match df with embeddings
          mask = responses.applymap(lambda x: pd.to_numeric(x, errors='coerce')).notna().all(axis=1)
          responses = responses[mask]
          data_s = data_s[mask]
          responses = responses.apply(pd.to_numeric)
          # print(pd.isnull(responses.apply(pd.to_numeric)).any().any()) # check if conversion failed
          responses = responses.apply(zscore, axis=0, nan_policy="omit") # zscore each questions

          #get ids
          situation_ids = responses.index         #get IDs of situations answered
          question_ids = responses.columns        #get IDs question answered for each situation

          X_stand = pd.DataFrame(StandardScaler().fit_transform(data_s),index=situation_ids) #standardize embeddings
          #With PCA, for regression, 90% variance explained
          X_pca = pd.DataFrame(PCA(0.95, random_state=0).fit_transform(X_stand),index=situation_ids)

          # import required data and labels
          list_par = getParams(m)

          ############################## 10-Fold cross validation:
          #Split dataframe 10 fold
          kf = KFold(n_splits=10, random_state=randState, shuffle=True)
          questions = list(kf.split(X_pca))
            
          ##################################### predictions #####################################

          for par in list_par:

            model, modelName = predModel(m, par, randState) 
            
            #Dataframes to store all predictions
            total_preds = pd.DataFrame(np.full(responses.shape, np.nan), columns=question_ids, index=situation_ids)

            for s_fold, fold_nr in zip(situations, range(len(situations))): #go through the question folds
                #print("Fold {}:".format(str(fold_nr + 1))) #status/progress
                
                #train/test index and embeddings for current fold
                sid_train = s_fold[0]
                sid_test = s_fold[1]
                s_test = X_pca.iloc[sid_test]
                
                # (train:) for each question go through all situations of the training fold
                for question, y in enumerate(responses.values.T):
                
                    s_train = X_pca.iloc[sid_train]
                    #get responses for ith question on the training situations
                    y_train = y[sid_train]
                    s_train = s_train.loc[~pd.isnull(y_train),:]
                    y_train = y_train[~pd.isnull(y_train)].astype('int')
                    
                    if (m==3) & (len(set(y_train)) == 1): #if only one type of response in training fold, use its value for prediction (because SVC does not work with only one class)
                      y_pred = np.repeat(y_train[0],s_test.shape[0])
                    else: 
                      y_pred = model.fit(s_train,y_train).predict(s_test)
                    
                    #save predictions in in dataframe
                    total_preds.iloc[sid_test, question] = y_pred
    
            # calculate and print performance
            results, resultsMean = corrQuestionBased(total_preds, responses)
            corr, pval, ci_lower, ci_upper, t, dof = resultsMean
            
            if verbose==1:
                #Model metrics:
                print("MODEL: " + modelName + " (par = " + str(par) + ") " + e.upper() + ":") #hand over string with specs...
                print("Correlation, lower-CI, upper-CI, p-value, t-statistic, DOF")
                print([round(corr,3), round(ci_lower,3), round(ci_upper,3), round(pval,3), round(t,3), round(dof,3)])
                print("\n")
            else:
                pass

            #save predictions
            total_preds.to_csv("../results/" + folder + "/Situations/" + d.upper() + "/" + modelName + "_" + str(par) + "_" + e.upper() + "_responses.csv") #save predictions -> to calculate and predict performance for plots
    return 1


def modelPerformance(m=0, par=1, d="RSQ", e="sentencebert", verbose=0, folder = "final"):
    
    # create folder to save results if not already
    Path("../results/" + folder + "/Situations/" + d.upper()).mkdir(parents=True, exist_ok = True)
    
    responses, data_s = loadData(data = d, folder = folder, emb = e)

    # drop all rows of responses that have is not interpretable as a number and match df with embeddings
    mask = responses.applymap(lambda x: pd.to_numeric(x, errors='coerce')).notna().all(axis=1)
    responses = responses[mask]
    data_s = data_s[mask]
    responses = responses.apply(pd.to_numeric)
    # print(pd.isnull(responses.apply(pd.to_numeric)).any().any()) # check if conversion failed
    responses = responses.apply(zscore, axis=0, nan_policy="omit") # zscore each questions

    #get ids
    situation_ids = responses.index         #get IDs of situations answered
    question_ids = responses.columns        #get IDs question answered for each situation

    X_stand = pd.DataFrame(StandardScaler().fit_transform(data_s),index=situation_ids) #standardize embeddings
    #With PCA, for regression, 90% variance explained
    X_pca = pd.DataFrame(PCA(0.95, random_state=0).fit_transform(X_stand),index=situation_ids)
    
    # import required data and labels
    list_par = getParams(m)
    
    ############################## 10-Fold cross validation:
    #Split dataframe 10 fold
    kf = KFold(n_splits=10, random_state=randState, shuffle=True)
    situations = list(kf.split(X_pca))

    #################################### predictions #####################################

    model, modelName = predModel(m, par, randState)
    
    #Dataframes to store all predictions
    total_preds = pd.DataFrame(np.full(responses.shape, np.nan), columns=question_ids, index=situation_ids)
    total_dumb = pd.DataFrame(np.full(responses.shape, np.nan), columns=question_ids, index=situation_ids)     #dumb predictor individual level

    for s_fold, fold_nr in zip(situations, range(len(situations))): #go through the situation folds
        if verbose>1:
            print("Fold {}:".format(str(fold_nr + 1))) #status/progress
        else:
            pass
            
        sid_train = s_fold[0]
        sid_test = s_fold[1]
        s_test = X_pca.iloc[sid_test]
        
        # (train:) for each question go through all situations of the training fold
        for question, y in enumerate(responses.values.T):
        
            s_train = X_pca.iloc[sid_train]
            #get responses for ith question on the training situations
            y_train = y[sid_train]
            s_train = s_train.loc[~pd.isnull(y_train),:]
            y_train = y_train[~pd.isnull(y_train)].astype('int')
            
            if (m==3) & (len(set(y_train)) == 1): #if only one type of response in training fold, use its value for prediction (because SVC does not work with only one class)
              y_pred = np.repeat(y_train[0],s_test.shape[0])
            else: 
              y_pred = model.fit(s_train,y_train).predict(s_test)
             
            #restrict to scale
            y_dumb = np.repeat(np.mean(y_train), s_test.shape[0])
            
            #save predictions in in dataframe
            total_preds.iloc[sid_test, question] = y_pred
            total_dumb.iloc[sid_test, question] = y_dumb
      
    results, resultsMean = corrQuestionBased(total_preds, responses)
    corr, pval, ci_lower, ci_upper, t, dof = resultsMean
    #naive baseline:
    results_dumb, resultsMean_dumb = corrQuestionBased(total_dumb, responses)
    corr_dumb, pval_dumb, ci_lower_dumb, ci_upper_dumb, t_dumb, dof_dumb = resultsMean_dumb
    
    if verbose>=1:
        print("MODEL: " + modelName + " (par = " + str(par) + ") " + e.upper() + ":") #hand over string with specs...
        print("Correlation, lower-CI, upper-CI, p-value, t-statistic, DOF")
        print([round(corr,3), round(ci_lower,3), round(ci_upper,3), round(pval,3), round(t,3), round(dof,3)])
        print("\n")
    else:
        pass

    total_preds.to_csv("../results/" + folder + "/Situations/" + d.upper() + "/" + modelName + "_" + str(par) + "_" + e.upper() + "_responses.csv")
    total_dumb.to_csv("../results/" + folder + "/Situations/" + d.upper() + "/baseline_" + e.upper() + "_responses.csv")
    return 1
```
